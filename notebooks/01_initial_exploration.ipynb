```python
df_clean = df_clean.drop(columns=['step', 'isFlaggedFraud'], axis='columns')
```

```python
# change column names
df_clean = df_clean.rename(columns = 
                     {'nameOrig' : 'sender_id', 
                      'oldbalanceOrg' : 'sender_old_balance', 
                      'newbalanceOrig': 'sender_new_balance', 
                      'nameDest' : 'reciever_id', 
                      'oldbalanceDest' : 'receiver_old_balance', 
                      'newbalanceDest': 'receiver_new_balance', 
                      'isFraud' : 'is_fraud'
                      }
                     )
```

```python
# convert columns to list and move reciever column beside the sender column 
df_columns = df_clean.columns.to_list()
new_position = 3

df_columns.insert(new_position, df_columns.pop(df_columns.index('reciever_id')))

df_clean = df_clean[df_columns]
```
